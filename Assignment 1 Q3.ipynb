```python
import numpy as np

# Create a 2D Numpy array of size 1x3 with elements of your choice
arr1 = np.array([[1, 2, 3]])

# Create a Numpy array of length 50 with zeroes as its elements
arr2 = np.zeros(50)

# Create a Numpy array of size 3x2 with elements of your choice
arr3 = np.array([[4, 5], [6, 7], [8, 9]])

# Multiply arr1 and arr3 using Numpy functions
arr4 = np.dot(arr1, arr3)

# Change 5th element of arr2 to a different number
arr2[4] = 10

if np.shape(arr4) == (1, 2) and arr2[4] != 0:
    print("Passed")
else:
    print("Fail")
```

```python
import numpy as np

# Create a 3x3 identity matrix I
I = np.eye(3)

# Create a 3x3 matrix with all elements as 1
ones_matrix = np.ones((3, 3))

# Compute the matrix 9I + 1
matrix_to_multiply = 9 * I + ones_matrix

# Perform the dot product of I and (9I + 1)
result = np.dot(I, matrix_to_multiply)

print("Result of the dot product:")
print(result)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.plot(xpoints, ypoints, 'o')  # 'o' specifies point markers without lines
plt.show()

# Plotting with marker: Plot these points with a star marker
plt.plot(xpoints, ypoints, '*')  # '*' specifies star markers
plt.show()

# Using fmt format, add circular marker, red color and dashed line
plt.plot(xpoints, ypoints, 'ro--')  # 'ro--' specifies red color, circle markers, and dashed line
plt.show()

# Add xlabel, ylabel and title for the plot
plt.plot(xpoints, ypoints, 'ro--')
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Plot with circular marker, red color, and dashed line")
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.scatter(xpoints, ypoints)
plt.show()

# Set color to the scatter plot. Blue, Green, Red and Yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.scatter(xpoints, ypoints, c=colors)
plt.show()
```

```python
import random
import numpy as np

# Set the seed of random to 20
random.seed(20)

arr1 = np.array([1, 24, 31, 45, 73, 81, 94, 25])

# Using the random module, pick 4 different random numbers from arr1 and return their sum
random_numbers = random.sample(list(arr1), 4)
sum_of_random_numbers = sum(random_numbers)

print("Randomly picked numbers:", random_numbers)
print("Sum of randomly picked numbers:", sum_of_random_numbers)
```
